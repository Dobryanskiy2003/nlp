```python
import pandas as pd
```

```python
data = pd.read_csv("Petitions.csv")

data = data.sample(1000, random_state=42)

data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
!pip install pymorphy3
import re

import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

from pymorphy3 import MorphAnalyzer
morph = MorphAnalyzer()

nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = stopwords.words('russian')
```

```python
stopwords
```

```python
def preprocessing(text):
    text_without_special_characters = re.sub('[^A-Za-zА-Яа-я]+', ' ', text)

    words = word_tokenize(text_without_special_characters)

    normal_forms = [morph.normal_forms(word)[0] for word in words]

    delete_stopwords = [word for word in normal_forms if word not in stopwords]

    return delete_stopwords
```

```python
X = [preprocessing(sentence) for sentence in data["public_petition_text"]]

X
```

```python
max_len = -1
for x in X:
    if max_len < len(x):
        max_len = len(x)

max_len
```

```python
from gensim.models import Word2Vec
from torch.nn.utils.rnn import pad_sequence
import torch

w2v = Word2Vec(sentences=X, min_count=1, vector_size=64, epochs=50)

X = [w2v.wv[x] for x in X]

X = pad_sequence([torch.tensor(x) for x in X], True)
X.shape
```

```python
le = LabelEncoder()
y = le.fit_transform(data["reason_category"])
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from torch.utils.data import TensorDataset, DataLoader
train_ds = TensorDataset(X_train, torch.from_numpy(y_train).type(torch.int64))
train_dl = DataLoader(train_ds, batch_size=32, shuffle=True)
```

```python
x_c, y_c = next(iter(train_dl))
x_c.shape, y_c.shape
```

```python
import torch.nn as nn
class Rnn(nn.Module):
    def __init__(self):
        super(Rnn, self).__init__()
        self.rnn = nn.RNN(64, 128, batch_first=True)
        self.first_linear = nn.Linear(128, 15)
    def forward(self, x):
        y = self.rnn(x)[1]
        y = self.first_linear(y[-1])
        return y
```

```python
loss = nn.CrossEntropyLoss()
model = Rnn()
optimizer = torch.optim.Adam(model.parameters(), lr=0.005)
```

```python
epochs = 10
for epoch in range(epochs):
    sum_loss = 0
    for x_b, y_b in train_dl:
        optimizer.zero_grad()
        outputs = model(x_b)
        loss_value = loss(outputs, y_b)
        sum_loss += loss_value
        loss_value.backward()
        optimizer.step()

    print(f'Эпоха {epoch + 1}, Значение функции потерь: {sum_loss/len(train_dl)}')
```

```python
answ = model.forward(X_test)
```

```python
answ_encoded = []
for i in answ:
    answ_encoded.append(torch.argmax(i))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, answ_encoded))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, answ_encoded)
```

```python
class LSTM(nn.Module):
    def __init__(self):
        super(LSTM, self).__init__()
        self.rnn = nn.LSTM(64, 64, batch_first = True)
        self.first_linear = nn.Linear(64, 15)
    def forward(self, x):
        y = self.rnn(x)[1][0]
        y = self.first_linear(y[-1])
        return y
```

```python
loss1 = nn.CrossEntropyLoss()
model1 = LSTM()
optimizer1 = torch.optim.Adam(model1.parameters(), lr=0.0025)
```

```python
epochs = 5
for epoch in range(epochs):
    sum_loss = 0
    for x_b, y_b in train_dl:
        optimizer1.zero_grad()
        outputs = model1(x_b)
        loss_value = loss1(outputs, y_b)
        loss_value.backward()
        optimizer1.step()
        sum_loss += loss_value

    print(f'Эпоха {epoch + 1}, Значение функции потерь: {sum_loss/len(train_dl)}')
```

```python
answ1 = model1.forward(X_test)

answ_encoded1 = []
for i in answ1:
    answ_encoded1.append(torch.argmax(i))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, answ_encoded1))
```

```python
accuracy_score(y_test, answ_encoded1)
```

```python
class GRU(nn.Module):
    def __init__(self):
        super(GRU, self).__init__()
        self.rnn = nn.GRU(64, 64, batch_first = True)
        self.first_linear = nn.Linear(64, 15)
    def forward(self, x):
        y = self.rnn(x)[1]
        y = self.first_linear(y[-1])
        return y
```

```python
loss2 = nn.CrossEntropyLoss()
model2 = GRU()
optimizer2 = torch.optim.Adam(model2.parameters(), lr=0.0025)
```

```python
epochs = 5
for epoch in range(epochs):
    for x_b, y_b in train_dl:
        optimizer2.zero_grad()
        outputs = model2(x_b)
        loss_value = loss2(outputs, y_b)
        loss_value.backward()

        optimizer2.step()

    print(f'Эпоха {epoch + 1}, Значение функции потерь: {loss_value.item()}')
```

```python
answ2 = model2.forward(X_test)
answ_encoded2 = []
for i in answ2:
    answ_encoded2.append(torch.argmax(i))
from sklearn.metrics import classification_report
print(classification_report(y_test, answ_encoded2))
```

```python
accuracy_score(y_test, answ_encoded2)
```
